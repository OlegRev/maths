4. Задание (в программе):
    Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), 
    
    взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def y(k,x):
    y = np.cos(k*x)
    return y
x = np.linspace(-10, 10, 222)

plt.plot(x, y(1,x))
plt.plot(x, y(3,x))

plt.title('График y(x) = cos(k*x)')

plt.xlabel('Переменная x')
plt.ylabel('Функция y(x)')
plt.legend(['k=1','k=3'])
plt.show()
```
